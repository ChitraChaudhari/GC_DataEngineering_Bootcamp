#**Data cleaning**

```python
tax_data.columns = tax_data.columns.str.replace(' ', '_').str.lower()
tax_data.info()
```

```python
tax_data.rename(columns={"adjusted_gross_income_(agi)":"agi","number_of_returns_with_total_income":"returns_with_total_income","number_of_returns_with_taxable_income":"returns_with_taxable_income"}, inplace=True)
tax_data.info()
```

#**Data analysis**

```python
correlation = tax_data.corr()
```

```python
import seaborn as sns

sns.heatmap(correlation, vmin=-1,vmax =1,center=0,cmap=sns.diverging_palette(200, 20, n=100))
```

Above heat map it shows there is a postive corelation between number of returns and total income 

```python
tax_data[["state","zipcode"]].nunique()
```

```python
df = tax_data.groupby(['state'])
state_df = df.sum()
state_df.drop(['zipcode','index'],axis=1,inplace=True)
state_df['avg_agi']= (state_df.agi/state_df.number_of_returns)
state_df['avg_total_income']= (state_df.total_income_amount/state_df.returns_with_total_income)
state_df['avg_taxable_income'] = (state_df.taxable_income_amount/state_df.returns_with_taxable_income)
state_df.head()
```
